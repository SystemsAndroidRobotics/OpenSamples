```python
#libraries
import pandas as pd #データフレームを扱うためのライブラリ読み込み
import numpy as np
import matplotlib.pyplot as plt

# データのパラメータ
N = 1055            # サンプル数
dt = 0.037          # サンプリング間隔[s]
fl = 0.5            # 下側カットオフ周波数
fh = 5.0            # 上側カットオフ周波数
t = np.arange(0, N*dt, dt) # 時間軸
freq = np.linspace(0, 1.0/dt, N) # 周波数軸

#input files
In0_name = 'log20181117185246_itoha_waist.csv' #入力ファイル名

#図のスタイル
plt.style.use('ggplot') #グラフの外観スタイル
plt.rcParams['font.family'] ='Times New Roman'
plt.rcParams['font.size'] = 8 #フォントの大きさ
```

```python
#データ読み込み
df = pd.read_csv(In0_name) 
f = df['dim1'].as_matrix()
```

```python
# 高速フーリエ変換（周波数信号に変換）
F = np.fft.fft(f)
  
# 正規化 + 交流成分2倍
F = F/(N/2)
F[0] = F[0]/2

# 配列Fをコピー
F2 = F.copy()

# バンドパス処理（カットオフ周波数範囲外の帯域と、右半分の帯域の周波数信号を0にする）
F2[(freq < fl)] = 0
F2[(freq > fh)] = 0
F2[(freq > 1/(dt*2))] = 0

# 高速逆フーリエ変換（時間信号に戻す）
f2 = np.fft.ifft(F2)

# 振幅を元のスケールに戻す
f2 = np.real(f2*N)
```

```python
# グラフ表示
plt.figure()
plt.figure(figsize=(8,6),dpi=100) #図全体のアスペクト比と解像度を設定して描画領域を確保
# 時間信号（元）
plt.subplot(221)
plt.plot(t, f, label='f(n)')
plt.xlabel("Time")
plt.ylabel("Signal")
plt.grid()
leg = plt.legend(loc=1)
leg.get_frame().set_alpha(1)

# 周波数信号(元)
plt.subplot(222)
plt.plot(freq, np.abs(F), label='|F(k)|')
plt.xlabel('Frequency')
plt.ylabel('Amplitude')
plt.grid()
leg = plt.legend(loc=1)
leg.get_frame().set_alpha(1)

# 時間信号(ハイパス処理後)
plt.subplot(223)
plt.plot(t, f2, label='f2(n)')
plt.xlabel("Time[s]")
plt.ylabel("Signal")
plt.grid()
leg = plt.legend(loc=1)
leg.get_frame().set_alpha(1)

# 周波数信号(ハイパス処理後)
plt.subplot(224)
plt.plot(freq, np.abs(F2), label='|F2(k)|')
plt.xlabel('Frequency[Hz]')
plt.ylabel('Amplitude')
plt.xlim([fl,fh])
plt.grid()
leg = plt.legend(loc=1)
leg.get_frame().set_alpha(1)
plt.show()
```
